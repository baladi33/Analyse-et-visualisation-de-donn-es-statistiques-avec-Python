```python
print(f"H0: \"les clients donnent les pourboires: 20% le jeudi, 10% le vendredi, 40% le samedi, 30% le dimanche\"")
print()

expeted_frequencies = np.array([0.2, 0.1, 0.4, 0.3])
expeted_frequencies = expeted_frequencies * len(df)
expeted_frequencies

observed_frequencies = df["day"].value_counts(normalize=False, sort=False).round(2).values

alpha = 0.02

p_value = chisquare(f_obs=observed_frequencies, f_exp=expeted_frequencies).pvalue

if p_value < alpha:
    print("Nous avons suffisamment d'évidences pour rejeter H0")
else:
    print("Nous n'avons pas suffisamment d'évidences pour rejeter H0")

p_value
```

```python
from scipy.stats import chisquare
import numpy as np

print("H0: les clients donnent les pourboires: 20% le jeudi, 10% le vendredi, 40% le samedi, 30% le dimanche")
print()
# Fréquences attendues (corrigé nom variable)
expected_frequencies = np.array([0.2, 0.1, 0.4, 0.3])
expected_frequencies = expected_frequencies * len(df)
# Fréquences observées (corrigé ordre + suppression round)
observed_frequencies = df["day"].value_counts().reindex(
    ["Thur", "Fri", "Sat", "Sun"]
).values
alpha = 0.02
p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies).pvalue
# Résultat
if p_value < alpha:
    print("Nous avons suffisamment d'évidences pour rejeter H0")
else:
    print("Nous n'avons pas suffisamment d'évidences pour rejeter H0")
p_value
```

```python
import scipy.stats as stats
from scipy.stats import ttest_1samp #test t de Student à un échantillon
alpha = 0.02
t_stat, p_value = stats.ttest_1samp(df["tip"], 3.5)
print("t-statistic:", t_stat)
print("p-value:", p_value)

# result
if p_value < alpha:
    print("On rejette H0")
else:
    print("On ne rejette pas H0")
```

```python
from scipy.stats import ttest_1samp

p_value = ttest_1samp(df["tip"], 3.5).pvalue

print("p-value:", p_value)
print("On rejette H0" if p_value < 0.02
       else "On ne rejette pas H0")
```
